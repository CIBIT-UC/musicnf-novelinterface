```python
import os

import pandas as pd

from src.my_settings import settings

sett = settings()
```

```python
# table data from BrainVoyager (.voi file: targets_prob-map.voi)
tab_file = os.path.join(sett["git_path"], "data", "targets_cog_table.txt")

# columns: name, mean x, mean y, mean z, std x, std y, std z, n voxels
# values are in mm (except n voxels)

# import table and add column names
df = pd.read_csv(tab_file, sep="\s+", header=None)
df.columns = [
    "name",
    "mean_x",
    "mean_y",
    "mean_z",
    "std_x",
    "std_y",
    "std_z",
    "n_voxels",
]

df
```

```python
# let's create a new column for the hemisphere (left or right) based on the name which as this format '13_rightPMC'
df["hemisphere"] = df["name"].str.split("_").str[-1]

# trim 'PMC' from the hemisphere
df["hemisphere"] = df["hemisphere"].str.replace("PMC", "")

# remove column 'name'
df = df.drop(columns=["name"])

df
```

```python
# I want an output table like this:
# ID x_left y_left z_left n_voxels_left x_right y_right z_right n_voxels_right
# the last two rows are the mean and STD values for each column

# let's create a new table with the desired format
df_out = pd.DataFrame(
    columns=[
        "SubjectID",
        "x_left",
        "y_left",
        "z_left",
        "n_voxels_left",
        "x_right",
        "y_right",
        "z_right",
        "n_voxels_right",
    ]
)

# add subject IDs
df_out["SubjectID"] = range(1, 23)

# add values for each column
df_out["x_left"] = df[df["hemisphere"] == "left"]["mean_x"].values
df_out["y_left"] = df[df["hemisphere"] == "left"]["mean_y"].values
df_out["z_left"] = df[df["hemisphere"] == "left"]["mean_z"].values
df_out["n_voxels_left"] = df[df["hemisphere"] == "left"]["n_voxels"].values

df_out["x_right"] = df[df["hemisphere"] == "right"]["mean_x"].values
df_out["y_right"] = df[df["hemisphere"] == "right"]["mean_y"].values
df_out["z_right"] = df[df["hemisphere"] == "right"]["mean_z"].values
df_out["n_voxels_right"] = df[df["hemisphere"] == "right"]["n_voxels"].values

# add mean and STD values for each column
df_out.loc[22] = [
    "mean",
    df_out["x_left"].mean(),
    df_out["y_left"].mean(),
    df_out["z_left"].mean(),
    df_out["n_voxels_left"].mean(),
    df_out["x_right"].mean(),
    df_out["y_right"].mean(),
    df_out["z_right"].mean(),
    df_out["n_voxels_right"].mean(),
]
df_out.loc[23] = [
    "std",
    df_out["x_left"].std(),
    df_out["y_left"].std(),
    df_out["z_left"].std(),
    df_out["n_voxels_left"].std(),
    df_out["x_right"].std(),
    df_out["y_right"].std(),
    df_out["z_right"].std(),
    df_out["n_voxels_right"].std(),
]

df_out
```

```python
# save table to the clipboard
df_out.to_clipboard(index=False, header=True)
```

# Probability Map

```python
left_pm_file = os.path.join(
    sett["bids_path"], "derivatives", "tbv-vois", "targets_prob-map_left.nii.gz"
)
right_pm_file = os.path.join(
    sett["bids_path"], "derivatives", "tbv-vois", "targets_prob-map_right.nii.gz"
)
```

```python
# sum the images
from nilearn import image

pm = image.math_img("img1 + img2", img1=left_pm_file, img2=right_pm_file)
```

```python
# display the left and right probability maps in the same plot - use plot_stat_map
import matplotlib.pyplot as plt
from nilearn import plotting

P = plotting.plot_stat_map(
    pm,
    display_mode="ortho",
    threshold=10,
    draw_cross=False,
    cmap="viridis",
    vmin=10,
    vmax=50,
)

# add color bar
colorbar = P._cbar
colorbar.set_label("overlap probability (%)")

# change figure size
plt.gcf().set_size_inches(8, 3)
```
