```python
import os
from src.my_settings import settings
from src.utils import apply_mask

sett = settings()
```

```python
task_run_labels = [
    "task-loc_run-1",
    "task-nf_run-1",
    "task-nf_run-2",
    "task-sham_run-1",
    "task-sham_run-2",
]
```

```python
# apply functional mask to each run of a subject

for ss in sett["sub_labels"]:
    for tt in task_run_labels:
        func_img_path = os.path.join(
            sett["derivatives_path"],
            f"sub-{ss}",
            "func",
            f"sub-{ss}_{tt}_space-MNI152NLin2009cAsym_desc-preproc_bold.nii.gz",
        )

        mask_img_path = os.path.join(
            sett["derivatives_path"],
            f"sub-{ss}",
            "func",
            f"sub-{ss}_{tt}_space-MNI152NLin2009cAsym_desc-brain_mask.nii.gz",
        )

        out_path = apply_mask(func_img_path, mask_img_path)
        print(f"Applied mask to {func_img_path} and saved to {out_path}")
```
