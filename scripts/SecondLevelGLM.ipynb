## Plots

```python
# plot Active
z_map_path = "/Volumes/T7/BIDS-MUSICNF/derivatives/nilearn-glm-group/group_task-nf_stat-z_con-MotorImageryMinusRest.nii.gz"
z_map_diff_thres, thres = threshold_stats_img(z_map_path, alpha=alpha, height_control=hc, cluster_threshold=ct)
print(f"Threshold: {thres}")

display = nlp.plot_glass_brain(z_map_diff_thres, threshold=thres, colorbar=True, plot_abs=False, display_mode='lyrz', title='Active NF', cmap='cold_hot')

# Add colorbar text
colorbar = display._cbar
colorbar.set_label('z-score (Bonferroni, p=0.05, k>10)')
```

```python
# plot Sham
z_map_path = "/Volumes/T7/BIDS-MUSICNF/derivatives/nilearn-glm-group/group_task-sham_stat-z_con-MotorImageryMinusRest.nii.gz"
z_map_diff_thres, thres = threshold_stats_img(z_map_path, alpha=alpha, height_control=hc, cluster_threshold=ct)
print(f"Threshold: {thres}")

display = nlp.plot_glass_brain(z_map_diff_thres, threshold=thres, colorbar=True, plot_abs=False, display_mode='lyrz', title='Sham NF', cmap='cold_hot', vmax=7.6)

# Add colorbar text
colorbar = display._cbar
colorbar.set_label('z-score (Bonferroni, p=0.05, k>10)')
```
